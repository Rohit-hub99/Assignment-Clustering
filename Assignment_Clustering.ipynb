# 1. What is unsupervised learning in the context of machine learning ?
  - Unsupervised learning is a type of machine learning where the model learns patterns from unlabeled data. It identifies structure, groupings, or features in the data without predefined outputs.

  - Clustering customers based on purchasing behavior without knowing their segments in advance.



# 2. How does K-Means clustering algorithm work ?
  - Choose the number of clusters K.
  - Initialize K centroids randomly.
  - Assign each data point to the nearest centroid (based on distance).
  - Update centroids by computing the mean of assigned points.
  - Repeat steps 3–4 until centroids stabilize or a maximum number of iterations is reached.

  - Clustering mall customers into 3 spending categories (low, medium, high) based on annual income and spending score.



# 3. Explain the concept of a dendrogram in hierarchical clustering ?
  - A dendrogram is a tree-like diagram that visualizes the arrangement of clusters formed by hierarchical clustering. It shows how individual data points merge into clusters step by step.

    - Vertical axis: Distance or dissimilarity between clusters.
    - Horizontal axis: Data points or clusters.
    - Cutting the dendrogram at a specific height determines the number of clusters.

  - In customer segmentation, a dendrogram can show how customers group together based on purchase behavior, helping decide the optimal number of clusters.

# 5. What are the advantages of DBSCAN over K-Means ?

# 7. What are the limitations of Hierarchical Clustering ?
  - Scalability: Not suitable for large datasets (high time & space complexity).
  - No reassignments: Once a merge/split is done, it can’t be undone.
  - Sensitive to noise and outliers.
  - Choice of linkage (e.g., single, complete) can affect results significantly.

  - With 10,000+ data points, hierarchical clustering becomes slow and memory-intensive, unlike K-Means or DBSCAN.





# 8. Why is feature scaling important in clustering algorithms like K-Means ?
  - Feature scaling is important in clustering (like K-Means) because it ensures all features contribute equally to the distance calculation.

     - K-Means uses Euclidean distance, which is sensitive to feature magnitude.

     - Without scaling, features with larger ranges dominate clustering.

  - If one feature is "age" (0–100) and another is "income" (in thousands), income will overly influence clustering unless both are scaled (e.g., using StandardScaler or MinMaxScaler).




# 9. How does DBSCAN identify noise points ?
  - DBSCAN identifies noise points as those that:
    
    - Have fewer than minPts neighbors within a given radius
    - Are not part of any cluster (i.e., not core points or reachable from core points).

    - These points are labeled as outliers.
  
  - In a 2D scatter plot, isolated points far from dense regions are marked as noise by DBSCAN.

     - Used to evaluate and compare clustering results.

   
  - After running K-Means, inertia helps assess if clusters are well-formed; a sudden drop in inertia when increasing

# 13. Can hierarchical clustering be used on categorical data ?
  - Yes, hierarchical clustering can be used on categorical data, but you must use an appropriate distance metric for categorical features, such as:

     - Hamming distance
     - Matching dissimilarity
     - Gower distance (for mixed data)

   - Standard Euclidean distance is not suitable for categorical data.

   - Clustering customer profiles with categorical features like gender, occupation, and region using hierarchical clustering with Hamming distance.




# 14. What does a negative Silhouette Score indicate ?
  - A negative Silhouette Score indicates that a data point is misclassified, meaning it is closer to points in another cluster than to points in its own cluster.

     - Overall negative average score suggests poor clustering quality.

  - If a customer is assigned to Cluster A but their behavior is more similar to Cluster B, their Silhouette Score may be negative.




# 15.  Explain the term "linkage criteria" in hierarchical clustering ?
  - Linkage criteria in hierarchical clustering define how to measure the distance between clusters when merging them.

     - Common types:
       - Single linkage: Minimum distance between points of two clusters.
       - Complete linkage: Maximum distance between points of two clusters.
       - Average linkage: Average distance between all pairs of points in two clusters.
       - Ward’s linkage: Minimizes variance within clusters after merging.

   - Using complete linkage tends to produce compact clusters by considering the farthest points between clusters.



# 16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?
  - K-Means performs poorly on varying cluster sizes/densities because:

    - It assumes clusters are spherical and similar-sized.
    - Uses distance to centroid, so large or dense clusters can dominate.
    - Struggles to separate clusters with different densities or irregular shapes.

  - K-Means may merge a small dense cluster into a larger sparse cluster, misclassifying points.




# 17.  What are the core parameters in DBSCAN, and how do they influence clustering /
  - Core parameters in DBSCAN:
     
     - ε (eps): Radius to search neighbors around a point.
        - Too small: many points labeled noise.
        - Too large: clusters may merge.

     - minPts: Minimum number of points required within
𝜀
ε to form a core point.
        - Low minPts: noise mistaken as clusters.
        - High minPts: small clusters ignored.



# 18.  How does K-Means++ improve upon standard K-Means initialization ?
  - K-Means++ improves K-Means by smarter initialization of centroids:
     
     - Selects the first centroid randomly.
     - Picks subsequent centroids probabilistically, favoring points far from existing centroids.
     - Results in better spread-out initial centroids.
     - Leads to faster convergence and improved clustering quality.

  - K-Means++ reduces chances of poor clusters caused by bad random initial centroids compared to standard K-Means.




# 19. What is agglomerative clustering ?
  - Agglomerative clustering is a type of hierarchical clustering that:
    
    - Starts with each data point as its own cluster.
    - Iteratively merges the closest clusters step-by-step.
    - Continues until all points form one cluster or a stopping criterion is met.

  - Grouping documents by similarity by repeatedly merging the two most similar clusters until a desired number of groups is reached.




# 20. What makes Silhouette Score a better metric than just inertia for model evaluation ?
  - Silhouette Score is better than inertia because:
    
    - It considers both intra-cluster cohesion and inter-cluster separation.
    - Works with any clustering algorithm.
    - Scales between -1 and 1, giving clear quality interpretation.
    - Inertia only measures compactness and favors more clusters.

  - Two clusterings with similar inertia can have very different Silhouette Scores, indicating one has better-defined clusters.

```python
# 1. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# Apply K-Means clustering with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Plot data points colored by cluster label and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X')
plt.title("K-Means Clustering with 4 Centers")
plt.show()
```

```python
# 23.  Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic moon-shaped data
X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Plot clusters: core points, border points, and outliers
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

for k, col in zip(unique_labels, colors):
    class_member_mask = (labels == k)
    if k == -1:
        # Outliers/noise plotted as red 'x'
        plt.scatter(X[class_member_mask, 0], X[class_member_mask, 1],
                    c='red', marker='x', label='Outliers')
    else:
        plt.scatter(X[class_member_mask, 0], X[class_member_mask, 1],
                    c=[col], label=f'Cluster {k}')

plt.title("DBSCAN Clustering with Outliers Highlighted")
plt.legend()
plt.show()
```

```python
# 24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Load Wine dataset
data = load_wine()
X = data.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means with 3 clusters (Wine has 3 classes)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Print size of each cluster
unique, counts = np.unique(labels, return_counts=True)
for cluster, size in zip(unique, counts):
    print(f"Cluster {cluster}: {size} samples")
```

```python
# 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# Calculate silhouette scores for k=2 to 5
k_values = range(2, 6)
scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Plot silhouette scores as bar chart
plt.bar(k_values, scores, color='skyblue')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for K-Means Clustering')
plt.xticks(k_values)
plt.show()
```

```python
# 30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage

from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Load Iris dataset
iris = load_iris()
X = iris.data

# Perform hierarchical clustering using average linkage
Z = linkage(X, method='average')

# Plot dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, labels=iris.target, leaf_rotation=90)
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.xlabel('Sample index or (cluster size)')
plt.ylabel('Distance')
plt.show()
```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate overlapping synthetic data
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)

# Fit K-Means with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Create mesh grid for decision boundaries
h = 0.1  # step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries and data points
plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=labels, s=30, cmap='viridis', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title("K-Means Clustering with Decision Boundaries")
plt.legend()
plt.show()
```

```python
# 43.  Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side

from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt

# Generate noisy circular data
X, _ = make_circles(n_samples=300, noise=0.05, factor=0.5, random_state=42)

# Apply KMeans (2 clusters)
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.15, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Plot side-by-side
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', s=30)
axs[0].set_title("KMeans Clustering")

axs[1].scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', s=30)
axs[1].set_title("DBSCAN Clustering")

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```

```python
# 44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt
import numpy as np

# Load Iris dataset
iris = load_iris()
X = iris.data

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Compute silhouette scores for each sample
sil_scores = silhouette_samples(X, labels)

# Sort scores within each cluster for better visualization
y_ticks = []
y_lower = 0
plt.figure(figsize=(8, 6))

for cluster in range(3):
    cluster_sil_scores = sil_scores[labels == cluster]
    cluster_sil_scores.sort()
    size = cluster_sil_scores.shape[0]
    y_upper = y_lower + size
    plt.barh(range(y_lower, y_upper), cluster_sil_scores, edgecolor='none', height=1)
    y_ticks.append((y_lower + y_upper) / 2)
    y_lower = y_upper + 10  # add gap between clusters

plt.xlabel('Silhouette Coefficient')
plt.ylabel('Samples')
plt.title('Silhouette Coefficient for Each Sample (Iris, KMeans)')
plt.yticks(y_ticks, [f'Cluster {i}' for i in range(3)])
plt.show()
```

```python
# 46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd

# Load Wine dataset
wine = load_wine()
X = wine.data[:, :4]  # first 4 features
feature_names = wine.feature_names[:4]

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Prepare DataFrame for seaborn
df = pd.DataFrame(X, columns=feature_names)
df['Cluster'] = labels.astype(str)

# Plot pairplot with cluster hue
sns.pairplot(df, hue='Cluster', palette='bright')
plt.suptitle("KMeans Clusters on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

```python
# 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate noisy blobs
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X)

# Count clusters (excluding noise) and noise points
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = np.sum(labels == -1)

print(f"Number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")
```
